```
complex_to_protein = pickle.load(open("/content/gdrive/My Drive/Bio 465 Research group/PICKLE_OBJECTS/BrentsComplexToProteins.pickle","rb"))
```

```
average_complex_variance = {}
for complex in complex_to_protein:
  added_proteins = 0
  protein_count = 0
  for protein in complex_to_protein[complex]:
    if (variance.index == protein).any():
      added_proteins = variance["variance_cancer_div_by_healthy"].loc[protein]
      protein_count += 1
  if protein_count == 0:
    protein_count += 1
  average = added_proteins / protein_count
  average_complex_variance[complex] = average
```
