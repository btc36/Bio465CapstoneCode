```
import pandas as pd
import numpy as np
import pickle
```

```
df = pickle.load(open("varianceDF.pickle",'rb')) #Messed up the file loading, sorry
```

```
df.head()
```

```
complex_to_protein = pickle.load(open("BrentsComplexToProteins.pickle","rb"))
```

```
stdChangeCoefficients = {}
for complexe in complex_to_protein:
  coefficients = []
  for protein in complex_to_protein[complexe]:
    if (df.index == protein).any():
      newCoefficient = df["variance_cancer_div_by_healthy"].loc[protein]
      coefficients.append(newCoefficient)
  
  sDev = np.nanstd(coefficients)
  stdChangeCoefficients[complexe] = sDev
  if sDev == None:
    print(complexe)
    print(coefficients)
    print("*****************")
```

```
print(np.nanstd(list(stdChangeCoefficients.values())))
print(np.nanmean(list(stdChangeCoefficients.values())))
print(list(stdChangeCoefficients.values()))
```

```
acv = open("/stdOfChangeCoefficientsPerComplex.pickle","wb")
pickle.dump(stdChangeCoefficients, acv)
```
