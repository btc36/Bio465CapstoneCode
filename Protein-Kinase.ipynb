```
import pandas as pd
import numpy as np
import pickle
import re
```

```
pkinfam_file = open("/content/gdrive/My Drive/Bio 465 Research group/pkinfam.txt", 'r')
pkins = set()
for line in pkinfam_file:
  line = line.strip()
  line = re.sub("\(","~",line)
  line = re.sub("\)","~",line)
  line = re.sub(" ","",line)
  line = line.split("~")
  pkins.add(line[1])
```

```
protein_to_complex = pickle.load(open("/content/gdrive/My Drive/Bio 465 Research group/PICKLE_OBJECTS/theRealProteinToComplexDict.pickle","rb"))
```

```
kinases = []
for protein in protein_to_complex:
  if protein in pkins:
    kinases.append(protein)
print(kinases)
print(len(kinases))
```

```
acv = open("/content/gdrive/My Drive/Bio 465 Research group/PICKLE_OBJECTS/kinases.pickle","wb")
pickle.dump(kinases, acv)
```
